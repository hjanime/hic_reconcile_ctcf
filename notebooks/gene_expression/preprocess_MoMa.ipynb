```python
import pandas as pd
import src
```

```python
gene_expression = pd.read_excel(src.external_data_path / "SupTab1_20180202_MoMaTA_RNAseq_DEseq2mmDIFF.xlsx")
```

```python
processed = gene_expression.copy()
processed = processed.drop("chr", axis=1)
processed = processed.rename(columns={'chr.1': 'chr'})
processed = processed[processed.gene.notnull()].reset_index(drop=True)
processed['strand'] = processed.strand.map(lambda x: '+' if x == 1 else "-")
processed = processed.sort_values(src.coords).reset_index(drop=True)
processed['start'] = processed.start.astype(int)
processed['end'] = processed.end.astype(int)
processed = processed[src.coords + [c for c in processed.columns if c not in src.coords]]
processed.head()
```

```python
processed.to_csv(src.processed_data_path / "gene_expression_MoMaTA_RNAseq.tsv", sep="\t", index=False, header=True)
```
