```python
import src
import pandas as pd
import numpy as np
```

```python
genes = pd.concat((pd.read_excel(src.external_data_path / '20190723_hg38TSSprotcodingnoYnoMT_4Luca.xlsx', sheet_name="+strand"),
                   pd.read_excel(src.external_data_path / '20190723_hg38TSSprotcodingnoYnoMT_4Luca.xlsx', sheet_name="-strand")),
                 axis=0)
genes.head()
```

```python
order = ['Chromosome/scaffold name', 'Transcription start site (TSS)', 'Transcription start site (TSS)', 'Strand']
order = order + [c for c in genes.columns if c not in order]
genes = genes[order]
genes['Chromosome/scaffold name'] = 'chr' + genes['Chromosome/scaffold name'].astype(str)
genes.columns = ['chr', 'tss_start', 'tss_end', 'strand'] + genes.columns[4:].tolist()
genes = genes.sort_values(['chr', 'tss_start', 'tss_end'])
genes['uid'] = np.arange(genes.shape[0], dtype=int)
genes['tss_end'] += 1
genes.head()
```

```python
genes[['chr', 'tss_start', 'tss_end', 'uid']].to_csv(src.interim_data_path / "hg38_ensembl_gene_tss.bed", sep='\t', index=False, header=False)
```

```python
hg19_genes = pd.read_csv(src.interim_data_path / "hg19_ensembl_gene_tss.bed", sep='\t', header=None, names = ['chr', 'tss_start', 'tss_end', 'uid'])
hg19_genes = hg19_genes.merge(genes.drop(['chr', 'tss_start', 'tss_end'], axis=1), on='uid')
hg19_genes.to_csv(src.interim_data_path / 'hg19_ensembl_gene_tss.tsv', sep='\t', index=False, header=True)
hg19_genes.head()
```
